####Multiple Linear Regression


```python
import numpy as np
import pandas as pd

df = pd.read_csv("diabetes.csv")

X = df[['Glucose', 'BMI', 'Age']].values
y = df['Outcome'].values.reshape(-1,1)

X = np.hstack((np.ones((X.shape[0],1)), X))

theta = np.linalg.inv(X.T @ X) @ X.T @ y

y_pred = X @ theta

mse = np.mean((y - y_pred) ** 2)
rmse = np.sqrt(mse)
r2 = 1 - (np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2))

print("Coefficients:", theta.flatten())
print("MSE:", mse)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("MLR: Actual vs Predicted")
plt.show()
```

##Polynomial Linear Regression

```python
import numpy as np
import pandas as pd

df = pd.read_csv("diabetes.csv")

X = df[['Glucose']].values
y = df['Outcome'].values.reshape(-1,1)

X_poly = np.hstack((
    np.ones((X.shape[0],1)),
    X,
    X**2
))

theta_poly = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y

y_poly_pred = X_poly @ theta_poly
mse_poly = np.mean((y - y_poly_pred) ** 2)
rmse_poly = np.sqrt(mse_poly)
r2_poly = 1 - (
    np.sum((y - y_poly_pred) ** 2) /
    np.sum((y - np.mean(y)) ** 2)
)

print("Polynomial MSE:", mse_poly)
print("Polynomial RMSE:", rmse_poly)
print("Polynomial R2:", r2_poly)
```

Plotting of Polynomial regression


```python
import matplotlib.pyplot as plt

sorted_idx = np.argsort(X.flatten())
X_sorted = X[sorted_idx]
y_poly_sorted = y_poly_pred[sorted_idx]

plt.scatter(X, y, label="Actual")
plt.plot(X_sorted, y_poly_sorted, label="Polynomial Fit")
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

comparison of Linear and Polynomial Regression

```python
print("Linear RMSE:", rmse)
print("Polynomial RMSE:", rmse_poly)

print("Linear R2:", r2)
print("Polynomial R2:", r2_poly)
```
