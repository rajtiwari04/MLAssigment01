```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/diabetes.csv")
data
```

outliers are highly sensitive to mean and  outliers are robust to median

Median is better when data contains outliers or is skewed, because it reflects the “typical” case more reliably.

```python
x=data['Age'].tolist()
y=data['BloodPressure']
```

```python
mean_x=sum(x)/len(x)
mean_y=sum(y)/len(y)

numerator=sum([(x[i]-mean_x)*(y[i]-mean_y) for i in range(len(x))])
denominator=sum([(x[i]-mean_x)**2 for i in range(len(x))])
```

```python
slope=numerator/denominator
slope
```

```python
intercept=mean_y-slope*mean_x
intercept
```

```python
def predict(x):
  return slope*x+intercept
```

```python
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, [predict(X) for X in x], color='red', label='Regression Line')
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.title("Simple Linear Regression (Glucose vs Outcome)")
plt.legend()
plt.show()
```

```python
ss_tot = sum([(Y - mean_y)**2 for Y in y])
ss_res = sum([(y[i] - predict(x[i]))**2 for i in range(len(x))])
r2 = 1 - (ss_res/ss_tot)

print("R-squared:", r2)
```
